```python
def add(a,b):
    return a+b
def sub(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /   ")
    if o == "+":
        print(add(a,b))
    elif o == "-":
        print(sub(a,b))
    elif o == "*":
        print(mul(a,b))
    elif o == "/" & b!=0:
        print(div(a,b))
    elif o == "/" & b ==0:
        print("The value of b can not be zero")
    else:
        print("Invalid operator")
        
cal()
```

```python
# 4! = 1*2*3*4
fac = 1
i = 1
n = int(input('Enter n = '))
def fact(n):
    global i,fac
    while i<=n:     
        fac = fac*i
        i = i+1
    return fac

print("The factorial = ",fact(n))
```

```python
# prime -> 2,3,5,7,11,13,17,19
# composite -> 4,6,8,9,10,12,15,16,18,20
i = 2
n = int(input("Enter n = "))
def cal(n):
    global i
    while i<n:
        if n%i ==0:
                print(n,"is composite no ")
                break
        i = i+1
    else:
        print(n,"is prime no")
    
cal(n)
```
